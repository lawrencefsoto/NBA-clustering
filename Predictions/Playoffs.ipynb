## Reading In My Data

```python
pg = pd.read_csv('../Positional clustering 17-18/pg_cluster_17_18')
sg = pd.read_csv('../Positional clustering 17-18/sg_cluster_17_18')
sf = pd.read_csv('../Positional clustering 17-18/sf_cluster_17_18')
pf = pd.read_csv('../Positional clustering 17-18/pf_cluster_17_18')
c = pd.read_csv('../Positional clustering 17-18/c_cluster_17_18')
```

## Cleaning Data apply some engineering

### PG

```python
pg.head() # sanity check
```

```python
## dummy out the position
pg_dummies = pd.get_dummies(pg['pg_cluster'])
pg_dummies
```

```python
## combining dummy and pg data
pg_concat = pd.concat([pg_dummies, pg], axis=1)
pg_concat.head()
```

```python
# setting index to Name, renaming dummies so I know what's what, dropping original pg_cluster after dummy.

pg_concat.set_index('Name', inplace=True)
pg_concat.rename(index=str, columns={0: 'pg_cluster_0', 1: 'pg_cluster_1', 2: 'pg_cluster_2'}, inplace=True)
pg_concat.drop(columns =['pg_cluster'], inplace = True)

# sorting the columns for pg. important to sort columns if you want to pickle your model for another use.
pg_concat = pg_concat.reindex(sorted(pg_concat.columns), axis=1)
pg_concat.head()
```

### SG

```python
sg.head()# sanity check
```

```python
sg_dummies = pd.get_dummies(sg['sg_cluster'])
sg_concat = pd.concat([sg_dummies, sg], axis=1)
sg_concat.set_index('Name', inplace=True)
sg_concat.rename(index=str, columns={0: 'sg_cluster_0', 1: 'sg_cluster_1'}, inplace=True)
sg_concat.drop(columns=['sg_cluster'], inplace=True)
sg_concat = sg_concat.reindex(sorted(sg_concat.columns), axis=1)

sg_concat.head()
```

### SF

```python
sf_dummies = pd.get_dummies(sf['sf_cluster'])
sf_concat = pd.concat([sf_dummies, sf], axis=1)
sf_concat.rename(index=str, columns={0: 'sf_cluster_0', 1: 'sf_cluster_1', 2:'sf_cluster_2'}, inplace=True)
sf_concat.drop(columns=['sf_cluster'], inplace=True)
sf_concat.set_index('Name', inplace=True)
sf_concat = sf_concat.reindex(sorted(sf_concat.columns), axis=1)
sf_concat.head()
```

### PF

```python
pf_dummies = pd.get_dummies(pf['pf_cluster'])
pf_concat = pd.concat([pf_dummies, pf], axis=1)
pf_concat.set_index('Name', inplace=True)
pf_concat.rename(index=str, columns={0: 'pf_cluster_0', 1: 'pf_cluster_1', 2:'pf_cluster_2', 3:'pf_cluster_3',
                                     4:'pf_cluster_4'}, inplace=True)
pf_concat.drop(columns=['pf_cluster'], inplace=True)
pf_concat = pf_concat.reindex(sorted(pf_concat.columns), axis=1)
pf_concat.head()
```

### C

```python
c_dummies = pd.get_dummies(c['c_cluster'])
c_concat = pd.concat([c_dummies, c], axis=1)
c_concat.set_index('Name', inplace=True)
c_concat.rename(index=str, columns={0: 'c_cluster_0', 1: 'c_cluster_1', 2:'c_cluster_2'}, inplace=True)
c_concat.drop(columns=['c_cluster'], inplace=True)
c_concat = c_concat.reindex(sorted(c_concat.columns), axis=1)
c_concat.head()
```

## Combining all the data together

```python
test = [pg_concat, sg_concat, sf_concat,pf_concat,c_concat]
season_17_18_cluster = pd.concat(test, sort=True)
season_17_18_cluster.head()
```

```python
test.isnull().sum() ## checking null values
```

```python
## Filling null values and just sorting again just to double check
season_17_18_cluster.fillna(0, inplace=True)
season_17_18_cluster = season_17_18_cluster.reindex(sorted(season_17_18_cluster.columns), axis=1)
```
