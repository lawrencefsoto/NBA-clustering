```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.basketball-reference.com/teams/GSW/2018.html'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
soup.find('div')
```

```python
soup.find('table', {'class': 'sortable stats_table'})
```

```python
ol = soup.find_all('div', {'id':'all_per_game'})
```

```python
soup.find('div', {'class':'overthrow table_container'}).find_all('td')
```

```python
table = soup.find('table', {'id': 'per_game'})
print(table)
```

```python
teams = []
for conference in ['E', 'W']:
    table = soup.find('table', {'id': 'confs_standings_' + conference}) ## this would be both E and W.
    #print(table)
    #print('')
    for row in table.find_all('tr')[1:]:  #[1:] skips first row
        team ={}
        team['Slug'] = row.find('a').text
        team['Name'] = row.find('a').attrs['title']
        team['Wins'] = row.find('td',{'data-stat': 'wins'}).text
        #team['Wins'] = row.find('td')[3].text
        team['Losses'] = row.find('td',{'data-stat': 'losses'}).text
        #team['Losses'] = row.find('td')[-1].text
        ## to make int you can wrap int()
        teams.append(team)
pd.DataFrame(teams)
```
